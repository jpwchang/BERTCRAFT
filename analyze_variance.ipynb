```python
import pandas as pd
import json
import os
```

```python
BASE_PATH = "/reef/bert_variance/{}/runs"
variance_wiki = []
for run in os.listdir(BASE_PATH.format("wikiconv")):
    with open(os.path.join(BASE_PATH.format("wikiconv"), run, "results.json")) as fp:
        variance_wiki.append(json.load(fp))
variance_wiki = pd.DataFrame.from_records(variance_wiki)
```

```python
variance_wiki = pd.concat([variance_wiki, pd.DataFrame(dict(variance_wiki.mean()), index=["Mean"]),pd.DataFrame(dict(variance_wiki.std()), index=["SD"])])
```

```python
variance_wiki
```

```python
print(variance_wiki.to_latex(float_format="{:.1%}".format))
```
